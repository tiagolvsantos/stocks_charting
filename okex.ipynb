```python
# Imports
import pandas as pd
from datetime import datetime, timedelta
import time, requests
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
def get_volume_oi(symbol: str):
    url = "{}/api/information/v3/{}/volume?granularity=86400".format("https://www.okex.com", symbol)
    df_volume_oi = pd.DataFrame(requests.get(url).json())
    df_volume_oi.columns  = ['Date', 'OI', 'Volume']
    return df_volume_oi

def get_top_traders_sentiment(symbol: str):
    url = "{}/api/information/v3/{}/sentiment?granularity=3600".format("https://www.okex.com", symbol)
    df_sentiment = pd.DataFrame(requests.get(url).json())
    df_sentiment.columns  = ['Date', 'Long', 'Short']
    return df_sentiment

def get_top_traders_long_short(symbol: str):
    url = "{}/api/information/v3/{}/long_short_ratio?granularity=86400".format("https://www.okex.com", symbol)
    df_long_short = pd.DataFrame(requests.get(url).json())
    df_long_short.columns  = ['Date', 'Long/Short Ratio']
    return df_long_short
```

```python
symbol="BTC"
```

```python
df_volume_oi = get_volume_oi(symbol)
df_sentiment = get_top_traders_sentiment(symbol)
df_long_short = get_top_traders_long_short(symbol)
```

```python
# Volume
fig = px.line(df_volume_oi , x="Date", y="Volume", title="{} Okex Volume".format(symbol), template = "plotly_dark")
fig.show()
```

```python
# Open Interest
fig = px.line(df_volume_oi , x="Date", y="OI", title="{} Okex Open Interest".format(symbol), template = "plotly_dark")
fig.show()
```

```python
# Top traders sentiment
fig = px.line(df_sentiment, x="Date", y=["Long","Short"], title='{} Okex  Top Traders Sentiment'.format(symbol), template = "plotly_dark")
fig.update_layout(legend_title="Side")
fig.show()
```

```python
# Top Traders Long Short
fig = px.line(df_long_short, x="Date", y="Long/Short Ratio", title='{} Okex  Top Traders Long / Short ratio'.format(symbol), template = "plotly_dark")
fig.show()
```
